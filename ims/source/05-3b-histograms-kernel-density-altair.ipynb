# Histograms

## Setup

```python
%matplotlib inline
import pandas as pd
import altair as alt
import numpy as np
```

## Normal distribution

### Create data

```python
# Create data with normal distribution
import numpy
import random

random.seed(123)

mean, sd = 11, 2
interest_rates_normal = numpy.random.normal(mean, sd, 1000)

df_normal = pd.DataFrame({"interest_rate": interest_rates_normal})
```

### Histogram

```python
# Histogram 
alt.Chart(df_normal).mark_bar().encode(
    alt.X("interest_rate:Q", bin=True),
    y='count()',
)
```

### Kernel density estimate plot

```python
 # Kernel density estimate (KDE) plot
alt.Chart(df_normal).transform_density(
    'interest_rate',
    as_=['interest_rate', 'density'],
).mark_area().encode(
    x="interest_rate:Q",
    y='density:Q',
)
```
