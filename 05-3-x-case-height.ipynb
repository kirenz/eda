# Case study: Height


## Setup

```python
%matplotlib inline
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import plotly.express as px

# Custom colors
blue = "#3F83F4"
blue_dark = "#062089"
blue_light = "#8DC0F6"
blue_lighter = "#BBE4FA"
grey = "#9C9C9C"
grey_dark = "#777777"
grey_light = "#B2B2B2"
orange = "#EF8733"
colors_blue = [blue, blue_light]
```

## Import data

```python
ROOT = "https://raw.githubusercontent.com/kirenz/modern-statistics/main/data/"
DATA = "height.csv"
df = pd.read_csv(ROOT + DATA)
```

```python
df["name"] = df["name"].astype("category")
df["gender"] = df["gender"].astype("category")
```

## Data exploration

```python
sns.pairplot(df, hue="gender");
```

```python
sns.histplot(data=df,  x="interest_rate", palette=colors_blue, binwidth=10)

plt.xlabel("Interest rate")
plt.ylabel("Count")

sns.despine();
```

```python
# Histogram with bin width of 30
sns.histplot(data=df,  x="interest_rate", palette=colors_blue, binwidth=30)

plt.xlabel("Interest rate")
plt.ylabel("Count")

sns.despine();
```

```python
# Histogram with bin width of 1
sns.histplot(data=df,  x="interest_rate", palette=colors_blue, binwidth=1)

plt.xlabel("Interest rate")
plt.ylabel("Count")

sns.despine();
```

```python
# Histogram 
sns.histplot(data=df,  x="interest_rate", palette=colors_blue, binwidth=2.5)

plt.xlabel("Interest rate")
plt.ylabel("Count")

sns.despine();
```

```python
# Histogram with additional styling
import matplotlib.ticker as ticker

sns.set_style("whitegrid", {'axes.grid' : False})
fig, ax = plt.subplots()

sns.histplot(data=df,  x= "interest_rate", palette=colors_blue)

ax.xaxis.set_major_formatter(ticker.EngFormatter('%'))
ax.yaxis.set_major_formatter(ticker.EngFormatter())
plt.xlabel("Interest rate")
plt.ylabel("Count")
sns.despine()

plt.show();
```
